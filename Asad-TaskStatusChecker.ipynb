Task - find status of wabsites (if the site is active or down) and store the information in a csv file and display it using pandas madule ;/

```python
import requests as req                 
import csv                             
import pandas as pd

header = ['URL', 'Status Code', 'Status']            # Header data for csv file
wabsites = []                                        # list to store wabsites urls in
while True:                                          # while this condition is true
    site = input("Enter Wabsite Name : ")            # take url from user to check status
    if site.lower() == 'done':                       # repeat untill user types done
        break
    else:
        wabsites.append(site)                        # append urls to wabsites list  

with open("asad.csv","w") as f:                      # open a new csv file for data to be stored In.
    wr = csv.writer(f,delimiter=",")                 # cammand that allow writing in csv file
    wr.writerow(header)                              # write header data to csv file

    for url in wabsites:                             # for every url list repeat the following
        data = req.get(url)                          # get data from site using request module
        status = data.status_code                    # extract status_code from data
        if status==200:                              # if status is 200 this means site is Active
            site_status = "Active"
        else:
            site_status = "Down"                     # else the wabsite is Down
        
        data = []                                    # created a new list to store folloing data
        data.append(url)                             # url of the site
        data.append(status)                          # status of the site
        data.append(site_status)                     # is the site Active or Down
        wr.writerow(data)                            # finally write the data to csv file

data= pd.read_csv("asad.csv")                        # Display csv data using pandas Library
data
```
